```python
import osmnx as ox
import geopandas as gpd
```

```python
# 1. Define the area of interest (Seattle and surrounding foothills)
place_name = "Seattle, Washington"
```

```python
# 2. Specify what we care about. 
# We'll use these as "proxies" for our Asset Value function.
tags = {
    'amenity': ['hospital', 'school', 'fire_station', 'police'],
    'building': ['apartments', 'commercial', 'industrial'],
    'landuse': ['residential', 'industrial']
}
```

```python
# 3. Download the geometries
# This gives you the polygons for every hospital, school, etc.
assets_gdf = ox.features_from_place(place_name, tags)
```

```python
# 4. Filter and simplify
# We only need the location and the type for now
assets_gdf = assets_gdf[['geometry', 'amenity', 'building', 'landuse']]
print(f"Successfully pulled {len(assets_gdf)} infrastructure assets.")
```

```python
assets_gdf.to_file("data/seattle_assets.geojson", driver="GeoJSON")
```
